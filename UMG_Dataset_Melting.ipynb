```r
master.data <- read.csv("UMG_Final_Tuzo.csv")
head(master.data)
```

```r
summary(master.data)
```

```r
length(names(master.data))
```

```r
volume.data <- master.data[,c(1,23:236)]
```

```r
revenue.data <- master.data[,c(1,237:450)]
```

```r
# Required for melt() and cast() function
install.packages("reshape2")
install.packages("reshape")

library(reshape2)
library(reshape)
```

```r
molten.volume.data <- melt(volume.data, id="UPC")
```

```r
head(molten.volume.data)
```

```r
unique(molten.volume.data$variable)
```

```r
names(molten.volume.data) <- c('UPC','week','volume')
```

```r
tail(molten.volume.data)
```

```r
molten.revenue.data <- melt(revenue.data, id='UPC')
tail(molten.revenue.data)
```

```r
combined.dataset <- cbind(molten.volume.data, molten.revenue.data$value)
head(combined.dataset)
tail(combined.dataset)
```

```r
# write.csv(combined.dataset,'tuzoUMGDataset.csv') 

activityDatesData <- master.data[,c(1,9)]

head(activityDatesData)

molten.activity.data <- melt(activityDatesData, id="UPC")

head(molten.activity.data)

combined.dataset <- cbind(combined.dataset, molten.activity.data$value)

head(combined.dataset)
```

```r
# combined.dataset <- read.csv('tuzoUMGDataset.csv')

sortedData <- combined.dataset[order(combined.dataset$UPC),]
tail(sortedData)

```

```r
head(sortedData)
```

```r
names(sortedData) <- c("UPC", "week","volume","revenue","activity.date")
cleanWeek <- function(week){
  substr(week,2,5)
}
```

```r
sortedData$week <- sapply(sortedData$week,FUN=cleanWeek)
head(sortedData)
```

```r
sortedData$week <- as.integer(sortedData$week)
sortedData$activity.date <- as.Date(sortedData$activity.date, "%d-%b-%y")
class(sortedData$activity.date)
```

```r
head(sortedData)
tail(sortedData)
dim(sortedData)
```

```r
write.csv(sortedData, "tuzoUMGDataset.csv")
```
