```r
validInvalidCalculation <- function(x,start_index,end_index){
  subsetted.data <- x[start_index:end_index,]
  length.subsetted.data <- length(subsetted.data$UPC)
  boolPrices <- !is.na(subsetted.data$price)
  for(trueValueIndex in 1:length.subsetted.data){
    if(boolPrices[trueValueIndex] == TRUE){
      start_index <- trueValueIndex
      break
    }
    else{
      subsetted.data[trueValueIndex,]$is.valid.product <- 'invalid'
      subsetted.data[trueValueIndex,]
    }
  }
  for(reverseid in length.subsetted.data:start_index){
    if(boolPrices[reverseid] == TRUE){
      break
    }
    else{
      subsetted.data[reverseid,]$is.valid.product <- "invalid"
    }
  }
  return(subsetted.data)
}



no.of.chunks <- length(sortedData$UPC)/228; 
# no.of.chunks <- 2; 
no.of.items.per.chunk <- 228

for(i in 0:(no.of.chunks - 1)){
  start <- i * no.of.items.per.chunk + 1
  end <- start + no.of.items.per.chunk - 1;

  print(validInvalidCalculation(sortedData, start, end))
}


```

```r
naList <- c(NA, NA, 1,2,2,3,NA,NA,4,4,NA,NA)

fillNaList <- function(x,firstBack = FALSE){
  lvls <- NULL
  if(is.factor(x)){
    lvls <- levels(x)
    x <- as.integer(x)
  }

  goodIdx <- !is.na(x)
  print(goodIdx)
  if(firstBack){
    goodVals <- c(x[goodIdx][1], x[goodIdx])
    print(goodVals)
  }
  else{
    goodVals <- c(NA, x[goodIdx])
    print(goodVals)
  }

  fillIdx <- cumsum(goodIdx)+1
  x<- goodVals[fillIdx]

  return(x)
}

print(fillNaList(naList, firstBack = TRUE))


traverseNA <- function(x){
  goodIdx <- !is.na(x)
  for(id in goodIdx){
    if(id){
      print('inside first loop')
      print(id)
      break
    }
  }
  for(reverseid in 12:1){
    if(goodIdx[reverseid] == TRUE){
      print('inside second loop')
      print(x[reverseid])
      break
    }
  }
}

traverseNA(naList)
```
